```python
# 필수 라이브러리 사용등록
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# `load_boston` has been removed from scikit-learn since version 1.2
# scikit-Learn 1.2버전이상에서는 load_boston이 없어짐
from sklearn.datasets import fetch_openml
```

### 분석모델 구축, 결과 분석
#### 전체 데이터(100)에서 보통 70~80%데이터를 훈련(train)시 사용, 20~30%데이터를 검증(test)시 사용

```python
## 회귀분석 모델, 함수 사용등록
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import sklearn
```

```python
# X(독립변수 13가지 속성들), y(종속변수, 독립변수의 영향받은 값)
y
```

```python
X
```

```python
# 통합데이터에서 독립변수만 다시 분리하려면[참조용]
X = dfBostonHousing.drop(['MEDV'], axis=1)
X
```

```python
# 훈련용 데이터와 평가(검증)용 데이터 분할 
# 판다스 DF에서 순서대로 데이터를 자르면, 고가의 집데이터가 후반부에 몰려있으면 훈련데이터와 검증데이터의 편차가 너무 심해짐(학습이 제대로 안됨)
# train_test_split() 데이터를 랜덤으로 잘라서 훈련데이터와 검증데이터를 분할 
# test_size=0.3 30% 0.2 조정가능, random_state=156, 105, 80 조정가능
# random_state=105 / RME 23 | 156 / 17.X | 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=156)
```

```python
X_test
```

```python
X_train
```

```python
X
```

```python
# 선형회귀 모델 생성
lr = LinearRegression()
```

```python
# 선형회귀 모델 훈련
lr.fit(X_train, y_train)
```

```python
X_test.info()
```

```python
# 선형회귀 분석 : 검증(평가)데이터를 넣어서 예측 수행
## NumPy배열로 변경하고 테스트할 것
y_predict = lr.predict(np.array(X_test))
```

```python
# 검증데이터 실제 가격값
y_test
```

```python
# 선형회귀모델로 예측값
y_predict
```

```python
## 제곱평균오차로 평가
mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse) # NumPy에 있는 squreroot 함수 사용하면 구해짐
r2 = r2_score(y_test, y_predict)

print(f'MSE = {mse}') # 거의 오차가 없음 
print(f'RMSE = {rmse}')
print(f'R^2(Variance score) = {r2}')
```

```python
## Y절편값, 회귀계수값
print(f'Y절편값 = {lr.intercept_}')
print(f'회귀계수값 = {np.round(lr.coef_, 1)}')
```

```python
pd.Series(data=np.round(lr.coef_, 2), index=X.columns)
```

```python
## 한글 사용 X
fig, axs = plt.subplots(figsize=(16, 16), ncols=3, nrows=4)

# category dType은 산점도를 그릴 수 없음
x_features = ['CRIM','ZN','INDUS','NOX','RM','AGE','DIS','TAX','PTRATIO','B','LSTAT']

# 배경색 설정
fig.set_facecolor('gray')  # 전체 플롯의 배경색 설정

# 각 서브플롯의 배경색 설정
for ax in axs.flatten():
    ax.set_facecolor('lightgrey')

for i, feat in enumerate(x_features) : # 독립변수 13개 이름으로 반복
    row = i // 3
    col = i % 3
    sns.regplot(data=dfBostonHousing, x=feat, y="MEDV", ax=axs[row][col], scatter_kws={'color':'black'},line_kws={'color':'white'})

    # x축과 y축의 글자색 변경
    axs[row][col].tick_params(axis='x', colors='white')  # x축 글자색 변경
    axs[row][col].tick_params(axis='y', colors='white')  # y축 글자색 변경


plt.show() # CHAS, RAD 형변환 전
```
