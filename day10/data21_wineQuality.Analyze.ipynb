##### statsmodes 라이브라리 분석 함수
- `OLS` >> Ordinary Least Square >> 모델로 분석을 하는 함수
- `GLM` >> Generalized Linear Model >> 일반 선형모델로 분석 함수
- `GLS` >> Generalized Least Square >> 2승 일반화된 최소 승수 모델 (두가지 표본집단을 나눠서)
- `WLS` >> Waited Least Square >> 가중치를 부가한 OLS 모델

- `fit()` >> 훈련결과를 내는 함수
- `predict()` >>

```python
# 필수 라이브러리/모듈 사용등록
import pandas as pd
from ucimlrepo import fetch_ucirepo # 인터넷으로 통계 데이터를 다운로드 하기 위한 함수
from scipy import stats # 과학수치 모듈 내의 기술통계 객체
from statsmodels.formula.api import ols, glm # 분석을 위한 결과를 도출해주는 함수
```

```python
## 와인 데이터 가져오기
wineQuality = fetch_ucirepo(id=186)
```

```python
dfWineTotal = wineQuality.data.original
```

```python
dfWineTotal.describe()
```

#### 회귀분석

```python
## 회귀분석 변수로 회귀수식 작성
regFormula = 'quality ~ fixed_acidity	+ volatile_acidity + citric_acid + residual_sugar +	chlorides +	free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'	
```

```python
## 회귀분석
regrResultOne = ols(regFormula, data=dfWineTotal).fit()
regrResultOne.summary()
```

```python
regrResultTwo = glm(regFormula, data=dfWineTotal).fit()
regrResultTwo.summary()
```

#### 도출된 회귀분석 결과로 새로운 와인샘플 품질등급 예측

```python
## 와인 품질 등급 예측용 샘플 데이터
dfWineTotal.columns.difference(['quality', 'color'])
```

```python
# 품질 예측이기 때문에 quality 불필요, 종류도 의미 없음
sampleDataOne = dfWineTotal[dfWineTotal.columns.difference(['quality', 'color'])]
```

```python
sampleDataOne = sampleDataOne[0:5]
```

```python
# OLS 모델 예측 시작
regrResultOne.predict(sampleDataOne)
```

#### 결론, 실제 데이터와 선형회귀 모델에서 예측한 결과가 거의 동일

```python
# 현재 실제 데이터중 와인 등급이 9인것
condition = dfWineTotal['quality'] == 9
dfWineTotal[condition]
```

```python
# 컬럼 순서를 바꿈, sampleDataOne으로 예측한 데이터와 컬럼순서를 일치시키기 위해
sampleDataTwo = pd.DataFrame(dummyData, columns=sampleDataOne.columns)
sampleDataTwo
```
