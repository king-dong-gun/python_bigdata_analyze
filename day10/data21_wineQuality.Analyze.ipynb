##### statsmodes 라이브라리 분석 함수
- `OLS` >> Ordinary Least Square >> 모델로 분석을 하는 함수
- `GLM` >> Generalized Linear Model >> 일반 선형모델로 분석 함수
- `GLS` >> Generalized Least Square >> 2승 일반화된 최소 승수 모델 (두가지 표본집단을 나눠서)
- `WLS` >> Waited Least Square >> 가중치를 부가한 OLS 모델

- `fit()` >> 훈련결과를 내는 함수
- `predict()` >>

```python
# 필수 라이브러리/모듈 사용등록
import pandas as pd
from ucimlrepo import fetch_ucirepo # 인터넷으로 통계 데이터를 다운로드 하기 위한 함수
from scipy import stats # 과학수치 모듈 내의 기술통계 객체
from statsmodels.formula.api import ols, glm # 분석을 위한 결과를 도출해주는 함수
```

```python
## 와인 데이터 가져오기
wineQuality = fetch_ucirepo(id=186)
```

```python
dfWineTotal = wineQuality.data.original
```

```python
dfWineTotal.describe()
```

#### 회귀분석

```python
## 회귀분석 변수로 회귀수식 작성
RFormula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides+ free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
## 회귀분석
regrResultOne = ols(RFormula, data=dfWineTotal).fit()
regrResultOne.summary()
```

```python
regrResultTwo = glm(RFormula, data=dfWineTotal).fit()
regrResultTwo.summary()
```

#### 도출된 회귀분석 결과로 새로운 와인샘플 품질등급 예측

```python
## 와인 품질등급 예측용 샘플데이터
dfWineTotal.columns.difference(['quality', 'color'])
```

```python
# 품질예측할것이기때문에 quality 불필요, 와인종류도 의미없음
sampleDataOne = dfWineTotal[dfWineTotal.columns.difference(['quality', 'color'])]
```

```python
sampleDataOne = sampleDataOne[0:5]
sampleDataOne
```

```python
dfWineTotal.head()
```

```python
# OLS 모델 예측 시작
regrResultOne.predict(sampleDataOne)
```

```python
# GLM 모델 예측
regrResultTwo.predict(sampleDataOne)
```

```python
## 시각화 라이브러리 사용등록
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm # 선형회귀 시각화를 위한 객체
```

```python
# 한글 꺠짐문제 해결
from matplotlib import rc

rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
# 레드와인 품질 가져오기
condition = dfWineTotal['color'] == 'red'
redWineQuality = dfWineTotal[condition]['quality']
```

```python
redWineQuality.info()
```

```python
# 화이트와인 품질 가져오기
# 레드와인 품질 가져오기
condition = dfWineTotal['color'] == 'white'
whiteWinQuality = dfWineTotal[condition]['quality']
```

```python
whiteWinQuality.info()
```

```python
# distplot() seaborn v0.14.0 에서 제거될 함수, 대신 histplot() 사용할 것
sns.histplot(redWineQuality, kde=True, color='red', label='레드와인')
sns.histplot(whiteWinQuality, kde=True, color='white', label='화이트와인')

plt.title('와인종류별 품질')
plt.legend()
plt.show()
```

```python
dfWineTotal.tail()
```

```python
others = list(set(dfWineTotal.columns).difference(set['quality', 'fixed_acidity']))
```

```python
# fixed_acidity(고정산)에대한 quality(품질) 의 상관관계를 선형회귀분석으로 시각화
sm.graphics.plot_partregress(exog_i='fixed_acidity', endog='quality', exog_others=others, data=dfWineTotal, obs_labels=False)

plt.show()
```

```python
# 선형회귀 결과 시각화
fig = plt.figure(figsize=(10, 14))
sm.graphics.plot_partregress_grid(regrResultOne, fig=fig)
plt.show()
```
