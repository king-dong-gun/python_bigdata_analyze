```python
# ì¥ì†Œê²€ìƒ‰ í•¨ìˆ˜ ìƒì„±
def find_place(shopName):
    # 1. ê²€ìƒ‰API ì£¼ì†Œ
    url = f'https://dapi.kakao.com/v2/local/search/keyword.json?query={shopName}'
    # 2. í—¤ë” êµ¬ì„±
    header = {'Authorization': 'KakaoAK b8a72979c0daa9b0e91e8ee94ceb77f4'}
    # 3. API ìš”ì²­ í›„ ì‘ë‹µë°›ìŒ
    places = requests.get(url, headers=header).json()['documents']
    # 4. í•„ìš” ì •ë³´ë§Œ ì¶”ì¶œ
    place = places[0]  # ì£¼ì†Œê°€ ì—¬ëŸ¬ê°œì¼ ë•Œ forë¬¸ì„ ëŒë ¤ì„œ ì œì£¼ë„ ì¸ê²ƒë§Œ ì¶”ì¶œí•´ì•¼í•¨
    realname = place['place_name']
    lat = place['y']  # ìœ„ë„(latitude)
    lng = place['x']  # ê²½ë„(longitude)
    phone = place['phone']
    realAddress = place['address_name']

    return [shopName, realname, phone, realAddress, lat, lng]
```

```python
# ì¥ì†Œê²€ìƒ‰ í•¨ìˆ˜ í…ŒìŠ¤íŠ¸
result = find_place('1100ê³ ì§€')
result
```

```python
# Locations ë¦¬ìŠ¤íŠ¸ë¥¼ forë¬¸ ëŒë ¤ì„œ ì¥ì†Œê²€ìƒ‰
locationResults = []
for loc in tqdm(Locations):
    try:
        locationResults.append(find_place(loc))
        time.sleep(0.1)
    except:
        pass
```

```python
dfLocationResult = pd.DataFrame(locationResults)
```

```python
# ì¤‘ë³µë°ì´í„° pivot
locationDataFinal = locationData.pivot_table(values='count', index=['ì‹¤ì œì´ë¦„', 'ì£¼ì†Œ', 'ìœ„ë„', 'ê²½ë„'], aggfunc='sum')
locationDataFinal.reset_index(inplace=True)
```

```python
locationDataFinal.to_excel('./data/3_locations_final.xlsx')
locationDataFinal = locationDataFinal.astype({'ìœ„ë„': 'float64', 'ê²½ë„': 'float64'})
```

```python
# êµ¬ê¸€ ì§€ë„ íƒ€ì¼ ì„¤ì •
tiles = "http://mt0.google.com/vt/lyrs=m&hl=ko&x={x}&y={y}&z={z}"
# ì†ì„± ì„¤ì •
attr = "Google"
# í•œë¼ì‚° ìœ„,ê²½ë„ ì„¤ì •
locHanlaMt = [33.362050, 126.530087]
map = folium.Map(location = locHanlaMt, zoom_start=11, tiles=tiles, attr=attr)

for i in range(len(locationDataFinal)): # 709ë²ˆ ë°˜ë³µ
    name = locationDataFinal['ì‹¤ì œì´ë¦„'][i] # DF columnsì´ë¦„ í–‰ê°’ == i
    count = int(locationDataFinal['count'][i])
    lat = float(locationDataFinal['ìœ„ë„'][i])
    lng = float(locationDataFinal['ê²½ë„'][i])
    folium.CircleMarker(location=[lat, lng], radius=count, color='red', popup=name).add_to(map)
map
```

```python
## ë§ˆì»¤ í´ëŸ¬ìŠ¤í„°
mapLocs = []
mapNames = []

for i in range(len(locationDataFinal)): # 709ë²ˆ ë°˜ë³µ
    data = locationDataFinal.iloc[i] # í–‰ì„ í•˜ë‚˜ì”© ì½ìŒ
    mapLocs.append([float(data['ìœ„ë„']), float(data['ê²½ë„'])]) # ìœ„ì¹˜ ì •ë³´ë¥¼ íŠœí”Œë¡œ ì¶”ê°€
    mapNames.append(data['ì‹¤ì œì´ë¦„'])

# ë§µ íƒ€ì¼ ë° ì†ì„± ì„¤ì •
tiles = "http://mt0.google.com/vt/lyrs=m&hl=ko&x={x}&y={y}&z={z}"
attr = "Google"

# ì œì£¼ë„ í•œë¼ì‚° ì£¼ë³€ ìœ„ì¹˜
locHanlaMt = [33.362050, 126.530087]
map = folium.Map(location = locHanlaMt, zoom_start=11, tiles=tiles, attr=attr)

# ë§ˆì»¤ í´ëŸ¬ìŠ¤í„° ìƒì„± ë° ì§€ë„ì— ì¶”ê°€
markerCluster = MarkerCluster(locations=mapLocs, popups=mapNames, name='ì œì£¼ í•«í”Œë ˆì´ìŠ¤', overlay=True, control=True)
markerCluster.add_to(map)
# ë ˆì´ì–´ ì»¨íŠ¸ë¡¤ ì¶”ê°€
folium.LayerControl().add_to(map)

map
```
