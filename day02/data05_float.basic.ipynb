````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
## 단일차트 >> 배열
plt.plot([1,2,3,4,5])
plt.show() # [<matplotlib.lines.Line2D at 0x10d7587a0>] 객체 정보 출력 없애기 >> 차트만 보기
````

````python
plt.plot(np.random.randn(50)) # np.random.randn(n) >> n개의 정수를 랜덤하게 차트 생성
plt.show()
````

````python
## 다중차트
fig = plt.figure()
cht1 = fig.add_subplot(2, 2, 1) # 1번 차트 2행 2열 생성
cht2 = fig.add_subplot(2, 2, 2) # 2번 차트 2행 2열 생성
cht3 = fig.add_subplot(2, 2, 3)
cht4 = fig.add_subplot(2, 2, 4)
# 차트는 4개까지 생성가능

cht1.plot(np.random.rand(50))
cht2.hist(np.random.rand(100))
cht3.scatter(np.arange(30), np.random.randn(30))
cht4.plot(np.random.randn(30),'violet') # 색상 'r', 'g', 'b', 'y', 'violet'

plt.show()
````

#### 차트 스타일 지정 옵션
- label, style, alpha, xticks, yticks, xlim, color, marker,...etc

````python
fig, axes = plt.subplots(2, 3, figsize=(18,6))
# color = 'k' >> 검정색, crimson >> 핏빛
# alpha = 0.1 >> 투명도 (min: 0.0, max: 1.0)
axes[0][0].plot(np.random.randn(50), color='crimson', alpha=1.0)
# bins = 30 >> 그루핑 갯수
axes[0][1].hist(np.random.randn(100), color='g', bins=30)
# linestyle = '-.' >> 선종류 ('-.', '--', '-', 'dotted', 'dashed', 'dashdot')
axes[0][2].plot(np.random.randn(100), color='k', linestyle='dashdot')
axes[1][0].scatter(np.arange(30), np.random.randn(30), marker='*')
axes[1][1].plot(np.random.randn(30),'violet')
axes[1][2].plot(np.random.randn(20), marker="o")

plt.show()
````

````python
# cumsum() cumulative sum(누적합)
# plot
plt.plot(np.random.randn(50).cumsum(), label='data1')
plt.plot(np.random.randn(50).cumsum(), label='data2')
plt.plot(np.random.randn(50), label='data3')
# legend() >> 범례표시
plt.legend(loc='upper left')
plt.show()
````

````python
## x, y 축 설정
plt.figure(figsize=(18, 5))
plt.plot(np.random.randn(1000))
plt.xticks([0, 250, 500, 750, 1000], labels=['min', '25%', 'median', '75%', 'max']) # x축 범위 설정
plt.yticks([-3, 0, 3]) # y축 범위 설정
plt.title('stastics') # 한글은 기본 지원X

plt.show()
````
