### 데이터 통합(~Day01)

#### merge
- 데이터를 합칠때 컬럼을 옆으로 늘려서 합침
- DataBase의 JOIN과 동일한 기능

```python
import pandas as pd
```

```python
# header 엑셀에서 필요없는 제목줄 등 없애기
# skipfooter 아래쪽에 필요없는 행들을 삭제
df_leftOne = pd.read_excel('../day01/data/sample_1.xlsx', header=1, skipfooter=2, usecols='A:D')
df_leftOne
```

```python
df_rightOne = pd.read_excel('../day01/data/sample_codemaster.xlsx')
df_rightOne
```

```python
## 병합 >> NaN 국적코드가 뭔지 몰라....
df_totalOne = pd.merge(left=df_leftOne, right=df_rightOne, 
                     how= 'left', left_on='국적코드', right_on='국적코드')
df_totalOne
```

```python
## DB INNER JOIN과 동일
df_totalOne = pd.merge(left=df_leftOne, right=df_rightOne,
                       how='inner', left_on='국적코드', right_on='국적코드')

# 일치하지 않는 값을 생략(삭제)
df_totalOne
```

### append
- 데이터를 합칠때 데이터 수가 늘도록 합치는 방법
- DB에 UNION이라는 기능과 완전동일

```python
df_leftTwo = pd.read_excel('./data/sample_2.xlsx', header=1, skipfooter=2, usecols='A:D')
df_leftTwo
```

```python
pd.__version__
```

```python
## df_all = df_total + df_total2
## 중요!! pandas 2버전 이상에서는 append() 함수가 없어짐
## pd.concat()으로 변경
df_all = df_leftTwo.append()
df_all
```

```python
pd.concat([df_leftOne, df_leftTwo])
```

```python
## 이렇게 합치면 안됨!!
pd.concat([df_totalOne, df_leftTwo])
```

```python
## df_left2를 가지고 dt_total2를 만듬. 국적명을 제대로 받기 위해서
df_totalTwo = pd.merge(left=df_leftTwo, right=df_rightOne,
                     how='left', left_on='국적코드', right_on='국적코드')
df_totalTwo
```

```python
# ignore_index는 DF의 자체 인덱스를 무시하고 새로 인덱스를 생성
df_all = pd.concat([df_totalOne, df_totalTwo], ignore_index=True)
df_all
```

```python
## DF 저장 pd.to_...
# index=False >> DF가 만드는 임의의 index는 생성X
df_all.to_excel('./data/df_all.xlsx', index=False)
```

#### Pivot 테이블
통계를 내면서 테이블을 세로로 된 DF를 가로로 바꿔주는 기능
DB에서 바로 피벗작업이 꽤 힘듦 >> 파이썬에서 피벗이 훨씬 쉬움

```python
## aggfunc >> sum(총합), mean(평균), min(최소), max(최대), median(중앙값), count(갯수), nunique(중복제거개수)
df_all.pivot_table(values='입국객수', index='국적코드', columns='기준년월', aggfunc='mean')
```
