```python
# í•„ìš” ë¼ì´ë¸ŒëŸ¬ë¦¬ ë“±ë¡
import pandas as pd
```

```python
raw_total = pd.read_excel('./data/1_crawling_raw.xlsx')
raw_total['tags']
```

```python
(raw_total['tags'][0][2:-2]).split("', '")
```

```python
# tqdm ëª¨ë“ˆì‚¬ìš©
from tqdm import tqdm
```

```python
tagAllList = []
count = 0
for tags in tqdm(raw_total['tags']):
    tagsList = tags[2:-2].split("', '")
    for tag in tagsList:
        tagAllList.append(tag)
```

```python
tagAllList
```

```python
# í•´ì‰¬íƒœê·¸ ì¶œí˜„ ë¹ˆë„
from collections import Counter
```

```python
# Counter í´ë˜ìŠ¤ë¡œ ê°™ì€ë¬¸ì 
tagCounts = Counter(tagAllList)
```

```python
tagCounts.most_common(50)
```

```python
# ë§›ì§‘ ì´ì™¸ì˜ í•´ì‹œíƒœê·¸ëŠ” ì œì™¸, ì›Œë“œí´ë¼ìš°ë“œ ê²°ê³¼ë³´ê³  ë‹¤ì‹œ í•„ìš”ì—†ëŠ” í•´ì‹œíƒœê·¸ ì¶”ê°€, ì •ì œí•œ ë’¤
STOPWORDS = ['#ì œì£¼ë„', '#ì œì£¼', '#jeju', '#ì œì£¼ì‹œ', '#ì¼ìƒ', '#ì œì£¼ë„ë¯¼', '#ë°˜ì˜êµ¬', '#ëˆˆì¹ë¬¸ì‹ ',
             '#ì œì£¼ìì—°ëˆˆì¹', '#ì œì£¼ì†ëˆˆì¹', '', '#ì„œê·€í¬', '#ì œì£¼ë„ê·¸ë¨', '#ì„ íŒ”', '#ì œì£¼ì‚´ì´', '#ì—¬í–‰ìŠ¤íƒ€ê·¸ë¨',
             '#ì œì£¼ë°˜ì˜êµ¬', '#ì œì£¼ëˆˆì¹ë¬¸ì‹ ', '#ì„œê·€í¬ëˆˆì¹ë¬¸ì‹ ', '#ì„œê·€í¬ë‚¨ìëˆˆì¹ë¬¸ì‹ ', '#ì„œê·€í¬ìì—°ëˆˆì¹',
             '#ì œì£¼ë‚¨ìëˆˆì¹ë¬¸ì‹ ', '#ì„œê·€í¬ì†ëˆˆì¹', '#ì„œê·€í¬ë°˜ì˜êµ¬', '#ì œì£¼ìŠ¤íƒ€ê·¸ë¨', '#ì†Œí†µ', '#ë§íŒ”', '#ì œì£¼ë©”ì´í¬ì—…',
             '#daily', '#ê°€ì¡±ì—¬í–‰', '#jejuisland', '#ì œì£¼ì†ëˆˆì¹ì—°ì¥', '#ì œì£¼í•´ì•ˆë„ë¡œ', '#ì…€ì¹´', '#ì œì£¼ê³µí•­',
             '#ì„œê·€í¬ë¬¸ì‹ ', '#ë°ì¼ë¦¬', '#ì œì£¼ë¬¸ì‹ ']

tagFilterList = []
for tag in tagAllList:
    if tag not in STOPWORDS:
        tagFilterList.append(tag)
```

```python
tagFilterCounts = Counter(tagFilterList)
tagFilterCounts.most_common(50)
```
